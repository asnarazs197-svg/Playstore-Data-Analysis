## Data Cleaning

```python
df['Installs'] = pd.to_numeric(df['Installs'].astype(str).str.replace(r'[^0-9.-]','',regex=True).replace('',None),errors='coerce')
df   #To correct the errors in Installs by replacing the different datatypes into float
```

```python
df['Price'] = pd.to_numeric(df['Price'].astype(str).str.replace(r'[^0-9.-]','',regex=True).replace('',None),errors='coerce')
df    #To correct the errors in Price by replacing the different datatypes into float
```

```python
df["Last Updated"] = pd.to_datetime(df["Last Updated"],errors="coerce")
df   #To correct the errors in Last updated by replacing the different datatypes into datetime 
```

## Converting size

```python
df.dtypes   #To show the datatypes 
```

- Function to convert size to MB

```python
def convert_size(size):
    if size == "Varies with device":
        return None
    elif "M" in size:
        return float(size.replace("M",""))
    elif "K" in size:
        return float(size.replace("K","")) / 1024
    else:
        return None

    
```

```python
df["Size_MB"] = df["Size"].apply(convert_size)
df  #Apply convertion
```

## Key Insights

📊 1. Most apps are free

- The majority of apps on the Play Store are free, while paid apps make up only a small percentage. This shows developers rely more on ads and in-app purchases than direct pricing.




⭐ 2. High ratings are common—but not always reliable
- Many apps have ratings between 4.0 and 4.5, but Apps with very few reviews can have misleadingly high ratings.
- Apps with more reviews give more trustworthy ratings





📂 3. Popular categories like:
    
- Family
- Game
- Tools

have the highest number of apps which means these areas are highly competitive.




📥 4. Most downloads come from a few apps

- A small number of apps (like social media or communication apps) dominate downloads, reaching millions or billions, while most apps have low installs.




💬 5. Reviews impact success

- Apps with more reviews tend to have higher installs 
- Be more visible and trusted




💰 6. Paid apps are less popular

- Paid apps generally have fewer installs
- Need strong value or uniqueness to succeed




📦 7. App size varies by category

- Games and multimedia apps are larger in size
- Tools and utility apps are smaller




🔄 8. Frequent updates improve ratings

- Apps that are updated regularly tend to have:

    Better user satisfaction

    Higher ratings




⚠️ 9. Some data inconsistencies exist

- The dataset may contain:

   Missing values

   Incorrect formats
So cleaning is important before analysis.




📈 10. Relationship between installs and ratings

- There is no strong direct relationship:

   Some highly downloaded apps have average ratings

   Some highly rated apps have low installs
